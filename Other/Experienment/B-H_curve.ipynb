```python
# 导入numpy和matplotlib库
import numpy as np
import matplotlib.pyplot as plt

# 从用户输入的数据中提取H和B列
H = np.array([0, 1.46, 2.92, 4.38, 5.84, 7.3, 10.95, 14.6, 18.25, 21.9, 29.2, 36.5])
B = np.array([0, 25.41, 50.82, 84.7, 114.95, 169.4, 229.9, 290.4, 338.8, 363, 387.2, 399.3])

# 使用三次样条插值来平滑B-H曲线
from scipy.interpolate import CubicSpline
cs = CubicSpline(H, B) # 创建三次样条插值对象
H_new = np.linspace(H.min(), H.max(), 1000) # 生成新的H值
B_new = cs(H_new) # 计算新的B值

# 绘制原始数据和平滑后的曲线
plt.scatter(H, B, label='raw data', color='blue') # 绘制散点图
plt.plot(H_new, B_new, label='curve', color='red') # 绘制曲线图
plt.xlabel('H (A/m)') # 设置x轴标签
plt.ylabel('B (T)') # 设置y轴标签
plt.title('B-H curve of the magnetic material') # 设置标题
plt.legend() # 显示图例
plt.show() # 显示图像
```

```python
# 导入numpy和matplotlib库
import numpy as np
import matplotlib.pyplot as plt

# 从用户输入的数据中提取H和B列
H = np.array([0, 1.46, 2.92, 4.38, 5.84, 7.3, 10.95, 14.6, 18.25, 21.9, 29.2, 36.5, 36.5, 29.2, 21.9, 18.25, 14.6, 10.95, 7.3, 5.84, 4.38, 2.92, 1.46, 0])
B = np.array([0, 25.41, 50.82, 84.7, 114.95, 169.4, 229.9, 290.4, 338.8, 363, 387.2, 399.3, 399.3, 387.2, 363, 338.8, 290.4, 229.9, 169.4, 114.95, 84.7, 50.82, 25.41, 0])

# 使用线性回归来拟合B-H曲线
from sklearn.linear_model import LinearRegression
lr = LinearRegression() # 创建线性回归对象
lr.fit(H.reshape(-1, 1), B) # 使用H和B进行拟合
B_new = lr.predict(H.reshape(-1, 1)) # 计算拟合后的B值

# 绘制原始数据和拟合后的曲线
plt.scatter(H, B, label='原始数据', color='blue') # 绘制散点图
plt.plot(H, B_new, label='拟合曲线', color='red') # 绘制曲线图
plt.xlabel('磁场强度 H (A/m)') # 设置x轴标签
plt.ylabel('磁通密度 B (T)') # 设置y轴标签
plt.title('B-H磁化曲线') # 设置标题
plt.legend() # 显示图例
plt.show() # 显示图像
```
