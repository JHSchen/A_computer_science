```python
import matplotlib.pyplot as plt
import numpy as np

# 实验数据
angles = np.array([20, 40, 60, 80, 100, 120])  # 散射角度
net_peak_areas = np.array([6895, 4988, 4145, 3399, 4175, 3986])  # 净峰面积

# 康普顿散射公式中的常数
h = 6.62607015e-34  # 普朗克常数 (焦耳·秒)
c = 2.99792458e8    # 光速 (米/秒)
moc2 = 511e3        # 电子静止质量能量 (电子伏特)
hv = 662e3          # 入射光子能量 (电子伏特)

# 计算理论微分截面的函数
def differential_cross_section(theta):
    theta_rad = np.radians(theta)  # 将角度转换为弧度
    lambda_c = h / (moc2 * c)  # 康普顿波长
    d_sigma_d_omega = (1 / (4 * np.pi)) * ((2 * np.pi * h**2) / (moc2**2 * c**4)) * \
                      ((1 + np.cos(theta_rad)**2) / (1 + (hv / moc2) * (1 - np.cos(theta_rad)))**2) * \
                      (1 + (hv / moc2)**2 * (1 - np.cos(theta_rad))**2 / \
                      ((1 + np.cos(theta_rad)**2) * (1 + (hv / moc2) * (1 - np.cos(theta_rad)))))
    return d_sigma_d_omega

# 计算理论微分截面
theoretical_differential_cross_sections = differential_cross_section(angles)

# 将实验值和理论值转换为百分比
max_net_peak_area = np.max(net_peak_areas)
experimental_percentages = (net_peak_areas / max_net_peak_area) * 100
theoretical_percentages = (theoretical_differential_cross_sections / np.max(theoretical_differential_cross_sections)) * 100

# 绘制百分比曲线
plt.figure(figsize=(10, 6))
plt.plot(angles, experimental_percentages, 'o-', label='实验值百分比')
plt.plot(angles, theoretical_percentages, 's-', label='理论值百分比')
plt.title('光子微分散射截面的实验值和理论值百分比')
plt.xlabel('散射角度 (度)')
plt.ylabel('微分散射截面百分比 (%)')
plt.legend()
plt.grid(True)
plt.show()

```

## B-Z Chemical ex

```python
import numpy as np
import matplotlib.pyplot as plt

# Parameters (you can adjust these)
k1 = 0.1  # Rate constant for the reaction
k2 = 0.09  # Rate constant for Ag+ reaction
Ag_initial = 0.9  # Initial Ag+ concentration

# Time and concentration arrays
t = np.linspace(0, 100, 1000)  # Time (arbitrary units)
BrO3 = np.zeros_like(t)  # BrO3- concentration
Ag = np.zeros_like(t)  # Ag+ concentration

# Initial conditions
BrO3[0] = 1.0  # Initial BrO3- concentration
Ag[0] = 0.9  # Initial Ag+ concentration

# Numerical simulation
for i in range(1, len(t)):
    # Reaction rates
    rate1 = k1 * BrO3[i - 1]
    rate2 = k2 * BrO3[i - 1] * Ag[i - 1]

    # Concentration changes
    BrO3[i] = BrO3[i - 1] - rate1
    Ag[i] = Ag[i - 1] + rate2

# Plot results
plt.figure(figsize=(8, 6))
plt.plot(t, BrO3, label="BrO3-")
plt.plot(t, Ag, label="Ag+")
plt.xlabel("Time")
plt.ylabel("Concentration")
plt.title("B-Z Oscillations")
plt.legend()
plt.grid(True)
plt.show()
```
