```python
# 导入sympy库和相关的函数
import sympy as sp
from sympy.physics.hydrogen import R_nl, E_nl
from sympy.plotting import plot

# 定义无量纲变量rho和主量子数n
rho = sp.Symbol("rho", positive=True)
n = sp.Symbol("n", positive=True, integer=True, nonzero=True)

# 定义原子序数Z和玻尔半径a0
Z = 1 # 氢原子
a0 = sp.Rational(52918, 100000000) # 单位是埃

# 定义一个函数，用于绘制径向波函数和概率密度函数
def plot_radial_wavefunction(n, l):
    # 计算归一化的径向波函数
    R = sp.simplify(R_nl(n, l, rho, Z))
    # 计算概率密度函数
    P = sp.simplify(rho**2 * R**2)
    # 计算能量
    E = sp.simplify(E_nl(n, Z))
    # 绘制图像
    p1 = plot(R, (rho, 0, 20), title=f"Radial wavefunction for n={n}, l={l}", xlabel=r"$\rho$", ylabel=r"$R_{nl}(\rho)$", show=False)
    p2 = plot(P, (rho, 0, 20), title=f"Probability density for n={n}, l={l}", xlabel=r"$\rho$", ylabel=r"$\rho^2 R_{nl}^2(\rho)$", show=False)
    p1.show()
    p2.show()
    # 打印相关信息
    print(f"For n={n}, l={l}:")
    print(f"Radial wavefunction: {R}")
    print(f"Probability density: {P}")
    print(f"Energy: {E}")

# 调用函数，绘制n=1,2,3时的图像
plot_radial_wavefunction(1, 0)
plot_radial_wavefunction(2, 0)
plot_radial_wavefunction(2, 1)
plot_radial_wavefunction(3, 0)
plot_radial_wavefunction(3, 1)
plot_radial_wavefunction(3, 2)
```
