```python
import matplotlib.pyplot as plt
import numpy as np
from scipy.optimize import curve_fit

# 数据
wavelengths = np.array([97, 286, 304, 500, 50, 100, 150, 200, 250, 300, 350, 400, 450])
reflectances = np.array([0.124535068, 0.023409588, 0.001944391, 0.14504, 0.00385306, 0.133032, 
                         0.118917, 0.827257, 0.357733, 0.0036922, 0.00257871, 0.00458544, 0.0912702])

# 定义拟合的函数形式
def func(x, a, b, c):
    return a * np.exp(-b * x) + c

# 初始参数猜测
initial_guess = (1.0, 0.001, 0.0)

# 拟合曲线
popt, pcov = curve_fit(func, wavelengths, reflectances, p0=initial_guess)

# 打印最优参数
print("Optimal parameters:", popt)

# 生成曲线上的点
x_curve = np.linspace(min(wavelengths), max(wavelengths), 500)
y_curve = func(x_curve, *popt)

# 绘图
plt.figure(figsize=(8, 6))
plt.scatter(wavelengths, reflectances, label='Data Points')
plt.plot(x_curve, y_curve, 'r-', label='Curve Fit')

plt.xlabel('Wavelength (nm)')
plt.ylabel('Reflectance (ReflectSHGfunc(φ))')
plt.title('Reflectance vs Wavelength with Curve Fit')

plt.legend()
plt.grid(True)
plt.tight_layout()

# 显示图形
plt.show()
```
