```python
import numpy as np
import matplotlib.pyplot as plt

def p(x):
    # Return the probability density function value for a given x
    a = 1 # Assume a = 1 for simplicity
    if abs(x) <= a:
        return 1 / (2 * a)
    else:
        return 0

a = 1 # Assume a = 1 for simplicity
x = np.arange(-2, 2, 0.04) # Generate x values from -2 to 2 with step size 0.04
y = [p(xi) for xi in x] # Generate y values by applying p(x) to each x value
plt.plot(x, y, lw=2) # Plot x and y with line width 2
plt.xlabel("x") # Set the x-axis label
plt.ylabel("p(x)") # Set the y-axis label
plt.grid(True) # Show the grid lines
plt.show() # Show the plot
```
