```python
# Import libraries
import numpy as np
import matplotlib.pyplot as plt

# Define constants
N = 100 # number of spins
mu = 1 # magnetic moment
k = 1 # Boltzmann constant
B = 1 # external magnetic field

# Define functions
def M(T):
  # magnetization per spin
  beta = 1/(k*T) # inverse temperature
  return mu*np.tanh(beta*mu*B)

def C(T):
  # heat capacity per spin
  beta = 1/(k*T) # inverse temperature
  return k*(beta*mu*B)**2*np.cosh(beta*mu*B)**(-2)

# Define temperature range
T = np.linspace(0.01, 10, 100) # avoid T = 0 to prevent division by zero

# Plot magnetization and heat capacity
plt.figure(figsize=(10,5))
plt.subplot(1,2,1)
plt.plot(T, M(T))
plt.xlabel('Temperature')
plt.ylabel('Magnetization per spin')
plt.title('Magnetization vs Temperature for B = {}'.format(B))
plt.subplot(1,2,2)
plt.plot(T, C(T))
plt.xlabel('Temperature')
plt.ylabel('Heat capacity per spin')
plt.title('Heat capacity vs Temperature for B = {}'.format(B))
plt.show()
```

```python
# Import libraries
import numpy as np
import matplotlib.pyplot as plt

# Define constants
N = 100 # number of spins
mu = 1 # magnetic moment
k = 1 # Boltzmann constant
B = 1 # external magnetic field

# Define function
def chi(T):
  # isothermal susceptibility per spin
  beta = 1/(k*T) # inverse temperature
  return (mu*B/k/T)**2*np.cosh(beta*mu*B)**(-2)

# Define temperature range
T = np.linspace(0.01, 10, 100) # avoid T = 0 to prevent division by zero

# Plot isothermal susceptibility
plt.figure()
plt.plot(T, chi(T))
plt.xlabel('Temperature')
plt.ylabel('Isothermal susceptibility per spin')
plt.title('Isothermal susceptibility vs Temperature for B = {}'.format(B))
plt.show()
```
