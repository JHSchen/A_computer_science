## Simulation of the motion and interaction dynamics Code

The algorithm for simulating the evolution of the model can be summarized by the following steps:

* The particles are placed at random in the center of the box with the constraint that no two particles can be closer than the length σ = 1 in the Lennard-Jones potential. This constraint prevents the initial force between any two particles from being too big, which would lead to the breakdown of the numerical method used to solve the differential equations. The velocity of each particle is assigned at random and then the velocity of the center of mass is set to zero.
* The total force on each particle is computed and the positions and velocities are updated according to the Verlet algorithm.
* The time is increased by the time step Δt.

### Main code

```python
# Refer to ThreePartsMD
# A program to simulate the motion and interaction of particles in a two-dimensional box


# import the necessary modules
import numpy as np
import random


# define the Lennard-Jones potential function
def LJ(r):
    # r: distance between two particles
    # return: potential energy between two particles
    return 4 * epsilon * ((sigma / r) ** 12 - (sigma / r) ** 6)

# define the Lennard-Jones force function
def LJ_force(r):
    # r: distance between two particles
    # return: force between two particles
    return -48 * epsilon * ((sigma / r) ** 13 - 0.5 * (sigma / r) ** 7)

# define the function to initialize the positions and velocities of the particles
def initialize():
    # return: pos, vel (initial positions and velocities of the particles)

    # initialize pos and vel as zero arrays
    pos = np.zeros((N, 2)) # pos[i] is the position vector of particle i
    vel = np.zeros((N, 2)) # vel[i] is the velocity vector of particle i

    # assign random positions to the particles in the middle region with no overlap
    n = 0 # number of assigned particles
    while n < N:
        x = random.uniform(1/3, 2/3) * L # random x coordinate in the middle region
        y = random.uniform(0, 1) * L # random y coordinate in the whole region
        r2min = np.min((pos[:n] - np.array([x, y])) ** 2, axis=1).sum() # minimum squared distance to other particles
        if n == 0 or r2min > sigma ** 2: # if no overlap
            pos[n] = np.array([x, y]) # assign position to particle n
            n += 1 # increase number of assigned particles

    # assign random velocities to the particles with zero center of mass velocity
    vel = np.random.randn(N, 2) # random velocity vectors from normal distribution
    vcm = vel.sum(axis=0) / N # center of mass velocity vector
    vel -= vcm # subtract center of mass velocity from each particle

    # return pos and vel
    return pos, vel

# define the function to update the positions and velocities of the particles using velocity Verlet algorithm
def update(pos, vel):
    # pos, vel: current positions and velocities of the particles
    # return: pos, vel (updated positions and velocities of the particles)

    # calculate the net force on each particle from Lennard-Jones potential
    force = np.zeros((N, 2)) # force[i] is the net force vector on particle i
    for i in range(N):
        for j in range(i + 1, N):
            rij = pos[i] - pos[j] # distance vector between particle i and j
            rij -= np.rint(rij / L) * L # apply periodic boundary condition
            r2 = rij ** 2 # squared distance between particle i and j
            fij = LJ_force(np.sqrt(r2)) * rij / np.sqrt(r2) # force vector between particle i and j from Lennard-Jones potential
            force[i] += fij # add force vector to net force on particle i
            force[j] -= fij # subtract force vector from net force on particle j

    # update positions and half-step velocities using velocity Verlet algorithm
    pos += vel * dt + 0.5 * force * dt ** 2 / m # update positions using current velocities and forces
    pos %= L # apply periodic boundary condition to positions
    vel += 0.5 * force * dt / m # update velocities by half-step using current forces

    # calculate the net force on each particle from Lennard-Jones potential at updated positions
    force = np.zeros((N, 2)) # force[i] is the net force vector on particle i
    for i in range(N):
        for j in range(i + 1, N):
            rij = pos[i] - pos[j] # distance vector between particle i and j
            rij -= np.rint(rij / L) * L # apply periodic boundary condition
            r2 = rij ** 2 # squared distance between particle i and j
            fij = LJ_force(np.sqrt(r2)) * rij / np.sqrt(r2) # force vector between particle i and j from Lennard-Jones potential
            force[i] += fij # add force vector to net force on particle i
            force[j] -= fij # subtract force vector from net force on particle j

    # update velocities by another half-step using updated forces
    vel += 0.5 * force * dt / m # update velocities using updated forces

    # return pos and vel
    return pos, vel

# define the function to simulate the motion and interaction of particles in the box
def simulate(n, p, t, dt):
    # n: initial number of particles in the middle region
    # p: probability of moving a particle from the middle region to either side
    # t: number of moves
    # dt: time step
    # return: positions, numbers (positions and numbers of particles in each region at each time step)

    # initialize positions and velocities of the particles
    pos, vel = initialize()

    # initialize numbers of particles in each region
    numbers = np.zeros((t + 1, 3), dtype=int) # numbers[i] is the number of particles in the left, middle and right region at time step i
    numbers[0, 1] = n # initial number of particles in the middle region

    # initialize positions of particles at each time step
    positions = [pos.copy()] # positions[i] is the position array of particles at time step i

    # loop for t times
    for i in range(t):
        # randomly choose a direction (left or right) and move a particle from the middle region to that direction with probability p
        direction = random.choice(["left", "right"])
        if random.random() < p:
            # find the indices of the particles in the middle region
            indices = np.where((pos[:, 0] > L / 3) & (pos[:, 0] < 2 * L / 3))[0]
            if len(indices) > 0: # if there are particles in the middle region
                # randomly choose one particle from the middle region
                index = random.choice(indices)
                # move it to the chosen direction by changing its x coordinate
                if direction == "left":
                    pos[index, 0] -= L / 3
                else:
                    pos[index, 0] += L / 3

        # update positions and velocities of the particles using velocity Verlet algorithm
        pos, vel = update(pos, vel)

        # update numbers of particles in each region by counting the x coordinates of the particles
        numbers[i + 1, 0] = np.count_nonzero(pos[:, 0] < L / 3) # number of particles in the left region
        numbers[i + 1, 1] = np.count_nonzero((pos[:, 0] > L / 3) & (pos[:, 0] < 2 * L / 3)) # number of particles in the middle region
        numbers[i + 1, 2] = np.count_nonzero(pos[:, 0] > 2 * L / 3) # number of particles in the right region

        # append positions of particles at this time step to the list
        positions.append(pos.copy())

    # return positions and numbers
    return positions, numbers
```

## Answer for prob1.1

* (a) 如果我们将N个粒子随机地放在盒子的中间部分，然后移除三个部分之间的“障碍”，粒子就会根据牛顿运动方程开始运动。这时，n1, n2, n3分别表示左、中、右部分中的粒子数，它们会随着时间t的变化而发生波动。一般来说，粒子会从中间部分向两边扩散，直到三个部分中的粒子数大致相等。这个过程表现出了时间的方向，即从有序到无序的方向。如果我们增加N的值，那么时间的方向就会更加明显，因为粒子数越多，从有序到无序的概率就越大。对于N=6，系统看起来没有明显的时间方向，因为粒子的数目在三个区域之间的波动很大，没有趋于平衡的迹象。可以看到Visible_Example2的fig非常明显。

* (b) 对于N=270，系统看起来有一个时间方向，因为粒子的数目在三个区域之间逐渐趋于平衡，即n1, n2, n3都接近N/3。当N增大时，这种趋势更加明显，而且达到平衡所需的时间也更长。多次运行程序，并选择不同的N值，观察n1, n2, n3随时间的变化曲线。们可以发现，当N增加时，这些特殊情况出现的概率会减小，所以我们对时间的方向的判断就会更加确定。

* (c) 如果我们在n1, n2, n3大致相等时反转时间，也就是让t变为-t，并且改变所有粒子的速度的符号，那么粒子会按照原来的轨迹返回到中间部分。这是因为牛顿运动方程具有时间反演对称性，也就是说如果把时间倒过来，物理规律不会改变。但是如果我们让粒子运动更长的时间再反转时间，那么粒子就不一定能回到中间部分了。这是因为粒子之间和盒子壁之间存在着碰撞和能量转移，这些过程会导致粒子速度和位置发生微小的变化。当时间反转时，这些微小的变化就会被放大，从而使得粒子偏离原来的轨迹。

* (d) 从粒子的运动可以看出，模拟中使用了周期性边界条件，也就是说当粒子从盒子的一边穿过时，它会从盒子的另一边进入。这样可以保证盒子内的粒子数不变，并且避免了边界效应。周期性边界条件是一种常用的模拟方法，在实际物理系统中也有对应的例子，例如晶体结构。在t=20时，那么时间方向就不太明显了。因为此时系统已经达到了平衡态，粒子的数目在三个区域之间只有微小的波动。无论正向还是反向播放视频，都看不出明显的差别。当N增大时，这种情况也不会改变。
  
* (e) 在图中显示的时间是以(m/ε)^1/2为单位的，其中σ和ε是Lennard-Jones势能(1.1)中的长度和能量参数，m是一个粒子的质量。你可以验证这个组合具有时间的单位，方法是将m用kg表示，将ε用J表示，将σ用m表示，然后计算(m/ε)^1/2的单位。对于氩气，σ=3.4×10^-10 m，ε=1.65×10^-21 J，m=6.69×10^-26 kg。氩气的(m/ε)^1/2的值是1.27×10^-12 s。在图中显示的时间为20时，真实的时间已经过去了20×(m/ε)^1/2=2.54×10^-11 s。

## Partical motion Code

```python
# 导入所需模块
import numpy as np
import matplotlib.pyplot as plt
import matplotlib.animation as animation

# 定义盒子大小和粒子数
box_size = 10 # 盒子边长
N = 32 # 粒子数

# 定义粒子类
class Particle:
    def __init__(self, x, y, vx, vy, r):
        self.x = x # x坐标
        self.y = y # y坐标
        self.vx = vx # x方向速度
        self.vy = vy # y方向速度
        self.r = r # 半径

    def move(self, dt):
        # 更新位置
        self.x += self.vx * dt
        self.y += self.vy * dt
        # 检查边界碰撞
        if self.x - self.r < 0 or self.x + self.r > box_size:
            self.vx = -self.vx # x方向反弹
        if self.y - self.r < 0 or self.y + self.r > box_size:
            self.vy = -self.vy # y方向反弹

    def collide(self, other):
        # 检查两个粒子之间是否发生碰撞
        dx = self.x - other.x
        dy = self.y - other.y
        dr = np.sqrt(dx**2 + dy**2)
        if dr < self.r + other.r:
            # 计算碰撞后的速度（假设弹性碰撞）
            dvx = self.vx - other.vx
            dvy = self.vy - other.vy
            nx = dx / dr
            ny = dy / dr
            dv = dvx * nx + dvy * ny
            self.vx -= dv * nx
            self.vy -= dv * ny
            other.vx += dv * nx
            other.vy += dv * ny

# 初始化粒子
particles = []
for i in range(N):
    # 随机生成位置和速度
    x = np.random.uniform(0, box_size)
    y = np.random.uniform(0, box_size)
    vx = np.random.uniform(-1, 1)
    vy = np.random.uniform(-1, 1)
    r = 0.2 # 统一半径
    p = Particle(x, y, vx, vy, r)
    particles.append(p)

# 定义时间步长和统计变量
dt = 1 # 时间步长
t = 0 # 当前时间
n_list = [] # 左半部分粒子数的列表
t_list = [] # 时间的列表

# 定义绘图相关的变量和函数
fig, (ax1, ax2) = plt.subplots(1, 2) # 创建两个子图
ax1.set_xlim(0, box_size) # 设置x轴范围
ax1.set_ylim(0, box_size) # 设置y轴范围
ax1.set_aspect('equal') # 设置等比例显示
ax1.set_title('Particle motion in a box') # 设置标题
ax1.axvline(box_size/2, color='k', linestyle='--') # 绘制中间分隔线
circles = [] # 存储每个粒子对应的圆形对象
for p in particles:
    c = plt.Circle((p.x, p.y), p.r, color='b') # 创建圆形对象
    ax1.add_patch(c) # 添加到子图中
    circles.append(c)

line, = ax2.plot([], [], color='r') # 创建折线对象
ax2.set_xlim(0, 100) # 设置x轴范围（时间）
ax2.set_ylim(0, N) # 设置y轴范围（粒子数）
ax2.set_xlabel('Time') # 设置x轴标签
ax2.set_ylabel('Number of particles on the left') # 设置y轴标签
ax2.set_title('Evolution of n(t)') # 设置标题

def init():
    # 初始化动画函数，返回需要更新的对象列表
    return circles + [line]

def update(frame):
    # 更新动画函数，参数frame表示帧数，返回需要更新的对象列表
    global t, n_list, t_list # 声明全局变量
    t += dt # 更新时间
    n = 0 # 计算左半部分粒子数
    for i in range(N):
        p = particles[i] # 获取第i个粒子对象
        p.move(dt) # 更新粒子位置
        circles[i].center = (p.x, p.y) # 更新圆形对象的中心坐标
        if p.x < box_size/2: # 如果粒子在左半部分盒子中
            n += 1 # 粒子数加一
    
    for i in range(N):
        for j in range(i+1, N):
            p1 = particles[i] # 获取第i个粒子对象
            p2 = particles[j] # 获取第j个粒子对象
            p1.collide(p2) # 检查并处理两个粒子之间的碰撞
    
    n_list.append(n) # 将当前左半部分粒子数添加到列表中
    t_list.append(t) # 将当前时间添加到列表中

    line.set_data(t_list, n_list) # 更新折线对象的数据

    return circles + [line]

# 创建动画对象，每隔interval毫秒更新一次，共运行frames帧，重复播放
ani = animation.FuncAnimation(fig, update, frames=5000, interval=100, init_func=init, repeat=True)

# 显示动画（如果需要保存动画，请取消下面一行的注释，并修改文件名和格式）
plt.show()
#ani.save('particles_in_box.mp4')
ani.save('demo.gif')
```

## Answer for prob1.2

(a) 这个问题是关于一个二维盒子中粒子的随机运动和碰撞的模拟。绘制了n(t)的变化曲线，其中n表示左半部分的粒子数。对于不同的N，也就是总粒子数，n(t)的行为会有所不同。一般来说，当N较小时，n(t)会有较大的波动，而当N较大时，n(t)会趋于平稳。这个系统会逐渐接近平衡状态，平衡状态的特征是n接近N/2，也就是左右两边的粒子数大致相等。平衡状态的定义随着N的增加而更加明确，因为当N增加时，n偏离N/2的概率会减小。平衡状态的定义不依赖于粒子最初在盒子中的分布方式，因为无论最初如何分布，粒子都会随机地向左右两边扩散,只要初始分布是随机的。

(b) 当系统达到平衡状态时，左半部分的粒子数n并不是一个常数，而是一个随机变量，它服从二项分布。二项分布的期望值是N/2，方差是Np(1-p)，其中p是每个粒子在左半部分的概率。由于粒子之间没有区别，所以p=0.5。因此，平衡状态下n的方差是N/4。平衡状态的本质是统计性质不随时间变化，而不是微观状态不变化。

(c) 如果N=32，那么系统有可能在某个时刻回到初始状态，也就是所有粒子都在盒子中间。这种情况出现的概率是(0.5)^32，约等于2.33e-10。要回到初始状态，需要所有粒子同时发生特定方向和速度的碰撞，这个概率非常小，但不是零。如果我们反转时间，也就是让t变为-t，并且改变所有粒子的速度的符号，那么系统一定会回到初始状态。这是因为牛顿运动方程具有时间反演对称性。

(d) n在平衡状态下的平均值与N成正比，也就是说n=N/2。如果我们从t=0开始计算平均值，那么我们得到的结果会受到初始条件的影响，并不能反映真正的平衡性质。为了得到正确的平衡平均值，我们需要等待系统达到平衡后再开始计算。清除之前的数据，并从当前时刻开始重新计算平均值。

(e) σ是n与其平均值之差的标准差。σ度量了n在平衡状态下的波动程度。如果n是一个常数，那么σ就是零。σ与N成反比关系，也就是说当N增加时，σ减小。σ/n与N成反比关系，也就是说当N增加时，σ/n减小。这意味着当N增加时，n更接近其期望值，平衡状态更加稳定。
