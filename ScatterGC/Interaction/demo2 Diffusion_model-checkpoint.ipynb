# 3D Diffusive Model for Spin Excitations in Metals

## Introduction
Spin excitations in metals are critical for understanding various physical phenomena and are essential in the field of spintronics. In a 3D metallic system, these excitations can be modeled using a diffusive approach, where the spin information propagates through the conduction electrons.

## Model Description
The model is based on the diffusion equation, which in three dimensions is given by:

$$
\frac{\partial s(\vec{r}, t)}{\partial t} = D \nabla^2 s(\vec{r}, t) - \frac{s(\vec{r}, t)}{\tau}
$$

where:
- $s(\vec{r}, t)$ is the spin excitation amplitude at position $\vec{r}$ and time $t$.
- $D$ is the diffusion constant.
- $\tau$ is the relaxation time.
- $\nabla^2$ is the Laplacian operator, representing diffusion in 3D space.

## Python Implementation


```python
import numpy as np
import matplotlib.pyplot as plt
from mpl_toolkits.mplot3d import Axes3D

# Constants
D = 0.1  # Diffusion constant (arbitrary units)
tau = 1e-12  # Relaxation time (s)
Lx, Ly, Lz = 1e-6, 1e-6, 1e-6  # System size (m)

# Discretization
dx = Lx / 100
dy = Ly / 100
dz = Lz / 100
dt = 0.01 * tau
Nx = int(Lx / dx)
Ny = int(Ly / dy)
Nz = int(Lz / dz)
T = int(10 * tau / dt)

# Initial condition: localized spin excitation at the center
s = np.zeros((Nx, Ny, Nz))
s[Nx//2, Ny//2, Nz//2] = 1

# Diffusion equation solver for 3D
for t in range(T):
    s_new = np.copy(s)
    for i in range(1, Nx-1):
        for j in range(1, Ny-1):
            for k in range(1, Nz-1):
                s_new[i, j, k] = s[i, j, k] + (D * dt / dx**2) * (
                    s[i+1, j, k] + s[i-1, j, k] + s[i, j+1, k] + s[i, j-1, k] + s[i, j, k+1] + s[i, j, k-1] - 6*s[i, j, k])
    s_new /= (1 + dt/tau)  # Relaxation term
    s = s_new

    if t % (T//10) == 0:
        # Create a figure and a 3D Axes
        fig = plt.figure()
        ax = fig.add_subplot(111, projection='3d')
        # Create a slice in the xy plane at z=Nz//2
        X, Y = np.meshgrid(np.linspace(0, Lx, Nx), np.linspace(0, Ly, Ny))
        Z = s[:, :, Nz//2]
        # Plot the surface.
        surf = ax.plot_surface(X, Y, Z, cmap='viridis')
        # Customize the z axis.
        ax.set_zlim(-0.1, 1.0)
        ax.zaxis.set_major_locator(plt.LinearLocator(10))
        ax.zaxis.set_major_formatter(plt.FormatStrFormatter('%.02f'))
        # Add a color bar which maps values to colors.
        fig.colorbar(surf, shrink=0.5, aspect=5)
        plt.title(f'Spin Excitation Amplitude at t={t*dt:.1e}s')
        plt.xlabel('x (m)')
        plt.ylabel('y (m)')
        plt.show()
```
