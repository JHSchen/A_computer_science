# E-H and SOC
## 1. Electron-Electron Interactions

Electron-electron interactions can be modeled using many-body perturbation theory, such as the GW approximation, where \( G \) is the Green's function and $W$ is the screened Coulomb interaction.

- **GW Approximation:**
  $ \Sigma_{e-e}(k, \omega) = i \int \frac{d^3q d\omega'}{(2\pi)^4} G(k-q, \omega-\omega') W(q, \omega')$
- **Self-Energy Function:**
  - Can be complex and often requires numerical methods to solve.
  - Can be approximated for certain limits (e.g., weak or strong coupling).

## 2. Spin-Orbit Coupling (SOC)

Spin-orbit coupling can split the electronic bands, leading to features that can be identified in ARPES spectra.

- **SOC Hamiltonian:**
  $\
  H_{SOC} = \lambda_{SOC} \mathbf{L} \cdot \mathbf{S}$
  where $\mathbf{L} $ is the orbital angular momentum and $\mathbf{S}$ is the spin angular momentum.
- **Effects:**
  - Causes band splitting.
  - Introduces additional complexity in the dispersion relations.

```python
import numpy as np
import matplotlib.pyplot as plt

# Define constants
hbar = 1.0545718e-34  # Planck's constant [J*s]
eV = 1.60218e-19      # Electron volt [J]
k_B = 1.380649e-23    # Boltzmann constant [J/K]

# Parameters for the simulation
omega_ph = 0.05       # Phonon energy [eV]
lambda_c = 1.0        # Electron-phonon coupling constant
v_F = 2.0             # Fermi velocity [eV * Å]
k_range = np.linspace(-1.0, 1.0, 500)  # Momentum range [1/Å]

# Bare electron dispersion (linear)
def bare_dispersion(k, v_F):
    return v_F * k

# Self-energy due to electron-phonon interaction
def self_energy(k, omega_ph, lambda_c):
    return lambda_c * omega_ph / (np.sqrt(k**2 + omega_ph**2))

# Renormalized dispersion
def renormalized_dispersion(k, v_F, omega_ph, lambda_c):
    return bare_dispersion(k, v_F) + self_energy(k, omega_ph, lambda_c)

# Calculate dispersions
bare_disp = bare_dispersion(k_range, v_F)
renorm_disp = renormalized_dispersion(k_range, v_F, omega_ph, lambda_c)

# Plot the results
plt.figure(figsize=(10, 6))
plt.plot(k_range, bare_disp, label='Bare Dispersion', linestyle='--')
plt.plot(k_range, renorm_disp, label='Renormalized Dispersion')
plt.axhline(y=omega_ph, color='gray', linestyle=':', label='Phonon Energy')
plt.xlabel('Momentum (k)')
plt.ylabel('Energy (E)')
plt.title('ARPES Dispersion with Electron-Phonon Coupling')
plt.legend()
plt.grid(True)
plt.show()
```

# Multiple interaction
1. **Eliashberg Function:**
   - Same as before, representing electron-phonon interaction.

2. **Electron-Electron Interaction:**
   - `self_energy_ee(k, lambda_ee)`: A simplified quadratic dependence on momentum. In reality, this can be more complex and require numerical integration.

3. **Spin-Orbit Coupling:**
   - `self_energy_soc(k, lambda_soc)`: A simplified model where SOC causes a linear splitting based on the sign and magnitude of momentum.

4. **Renormalized Dispersion:**
   - `renormalized_dispersion(k, v_F, omega_ph, lambda_ph, lambda_ee, lambda_soc)`: Combines the bare dispersion with all self-energy terms to produce the final renormalized dispersion.

```python

import numpy as np
import matplotlib.pyplot as plt

# Define constants
hbar = 1.0545718e-34  # Planck's constant [J*s]
eV = 1.60218e-19      # Electron volt [J]
k_B = 1.380649e-23    # Boltzmann constant [J/K]

# Parameters for the simulation
omega_ph = 0.05       # Phonon energy [eV]
lambda_ph = 1.0       # Electron-phonon coupling constant
lambda_ee = 0.1       # Electron-electron coupling constant (simplified)
lambda_soc = 0.02     # Spin-orbit coupling constant
v_F = 2.0             # Fermi velocity [eV * Å]
k_range = np.linspace(-1.0, 1.0, 500)  # Momentum range [1/Å]

# Bare electron dispersion (linear)
def bare_dispersion(k, v_F):
    return v_F * k

# Eliashberg function for electron-phonon interaction
def eliashberg_function(omega, omega_ph, lambda_ph):
    return lambda_ph * (omega_ph ** 2) / (omega ** 2 + omega_ph ** 2)

# Self-energy due to electron-phonon interaction (real part)
def self_energy_ph(k, omega_ph, lambda_ph):
    omega = bare_dispersion(k, v_F)
    return -eliashberg_function(omega, omega_ph, lambda_ph)

# Self-energy due to electron-electron interaction (simplified)
def self_energy_ee(k, lambda_ee):
    return -lambda_ee * bare_dispersion(k, v_F) ** 2

# Self-energy due to spin-orbit coupling (simplified)
def self_energy_soc(k, lambda_soc):
    return lambda_soc * np.sign(k) * np.sqrt(np.abs(k))

# Renormalized dispersion
def renormalized_dispersion(k, v_F, omega_ph, lambda_ph, lambda_ee, lambda_soc):
    return (bare_dispersion(k, v_F)
            + self_energy_ph(k, omega_ph, lambda_ph)
            + self_energy_ee(k, lambda_ee)
            + self_energy_soc(k, lambda_soc))

# Calculate dispersions
bare_disp = bare_dispersion(k_range, v_F)
renorm_disp = renormalized_dispersion(k_range, v_F, omega_ph, lambda_ph, lambda_ee, lambda_soc)

# Plot the results
plt.figure(figsize=(10, 6))
plt.plot(k_range, bare_disp, label='Bare Dispersion', linestyle='--')
plt.plot(k_range, renorm_disp, label='Renormalized Dispersion')
plt.axhline(y=omega_ph, color='gray', linestyle=':', label='Phonon Energy')
plt.xlabel('Momentum (k)')
plt.ylabel('Energy (E)')
plt.title('ARPES Dispersion with Multiple Interactions')
plt.legend()
plt.grid(True)
plt.show()
```

# superconducting gap and the formation of Cooper pairs with travel electron
## Key Concepts

1. **Superconducting Gap $ \Delta_{SC}$**:
   - In a superconductor, a gap opens at the Fermi level due to the formation of Cooper pairs.
   - This gap is an essential signature of superconductivity and appears as a suppression of electronic states at the Fermi level in the ARPES spectra.

2. **Bogoliubov Quasiparticles**:
   - The excitations in a superconductor are called Bogoliubov quasiparticles, which are superpositions of electron and hole states.
   - The dispersion of these quasiparticles reflects the opening of the superconducting gap and can be described by the Bogoliubov-de Gennes (BdG) equations.

3. **Self-Energy Modifications**:
   - In the superconducting state, the self-energy due to electron-phonon and electron-electron interactions needs to be reconsidered, as these interactions contribute to the pairing mechanism and the superconducting gap.

## Model

1. **Superconducting Gap $\Delta_{SC}$:**
   - `delta_sc = 0.02`: The superconducting gap value, representing the energy required to break a Cooper pair.

2. **Bogoliubov Quasiparticle Dispersion:**
   - `bogoliubov_dispersion(k, v_F, delta_sc)`: Calculates the dispersion of Bogoliubov quasiparticles in the superconducting state. The formula $ \sqrt{\epsilon_k^2 + \Delta_{SC}^2}$ reflects the opening of the superconducting gap.

3. **Renormalized Dispersion in the Superconducting State:**
   - `renormalized_dispersion_sc(k, v_F, omega_ph, lambda_ph, lambda_ee, lambda_soc, delta_sc)`: Combines the bare dispersion, self-energy terms, and the superconducting gap to give the final renormalized dispersion in the superconducting state.

4. **Calculations and Plotting:**
   - The code calculates the bare dispersion, Bogoliubov quasiparticle dispersion, and renormalized dispersion in the superconducting state over the momentum range.
   - The plot shows how the combined interactions modify the electronic dispersion, highlighting the superconducting gap and the effects of electron-phonon, electron-electron, and spin-orbit interactions.

```python
import numpy as np
import matplotlib.pyplot as plt

# Define constants
hbar = 1.0545718e-34  # Planck's constant [J*s]
eV = 1.60218e-19      # Electron volt [J]
k_B = 1.380649e-23    # Boltzmann constant [J/K]

# Parameters for the simulation
omega_ph = 0.05       # Phonon energy [eV]
lambda_ph = 1.0       # Electron-phonon coupling constant
lambda_ee = 0.1       # Electron-electron coupling constant (simplified)
lambda_soc = 0.02     # Spin-orbit coupling constant
delta_sc = 0.02       # Superconducting gap [eV]
v_F = 2.0             # Fermi velocity [eV * Å]
k_range = np.linspace(-1.0, 1.0, 500)  # Momentum range [1/Å]

# Bare electron dispersion (linear)
def bare_dispersion(k, v_F):
    return v_F * k

# Eliashberg function for electron-phonon interaction
def eliashberg_function(omega, omega_ph, lambda_ph):
    return lambda_ph * (omega_ph ** 2) / (omega ** 2 + omega_ph ** 2)

# Self-energy due to electron-phonon interaction (real part)
def self_energy_ph(k, omega_ph, lambda_ph):
    omega = bare_dispersion(k, v_F)
    return -eliashberg_function(omega, omega_ph, lambda_ph)

# Self-energy due to electron-electron interaction (simplified)
def self_energy_ee(k, lambda_ee):
    return -lambda_ee * bare_dispersion(k, v_F) ** 2

# Self-energy due to spin-orbit coupling (simplified)
def self_energy_soc(k, lambda_soc):
    return lambda_soc * np.sign(k) * np.sqrt(np.abs(k))

# Bogoliubov quasiparticle dispersion in the superconducting state
def bogoliubov_dispersion(k, v_F, delta_sc):
    epsilon_k = bare_dispersion(k, v_F)
    return np.sqrt(epsilon_k**2 + delta_sc**2)

# Renormalized dispersion in the superconducting state
def renormalized_dispersion_sc(k, v_F, omega_ph, lambda_ph, lambda_ee, lambda_soc, delta_sc):
    epsilon_k = bare_dispersion(k, v_F)
    self_energy = (self_energy_ph(k, omega_ph, lambda_ph)
                   + self_energy_ee(k, lambda_ee)
                   + self_energy_soc(k, lambda_soc))
    return np.sqrt((epsilon_k + self_energy)**2 + delta_sc**2)

# Calculate dispersions
bare_disp = bare_dispersion(k_range, v_F)
bogoliubov_disp = bogoliubov_dispersion(k_range, v_F, delta_sc)
renorm_disp_sc = renormalized_dispersion_sc(k_range, v_F, omega_ph, lambda_ph, lambda_ee, lambda_soc, delta_sc)

# Plot the results
plt.figure(figsize=(10, 6))
plt.plot(k_range, bare_disp, label='Bare Dispersion', linestyle='--')
plt.plot(k_range, bogoliubov_disp, label='Bogoliubov Quasiparticles')
plt.plot(k_range, renorm_disp_sc, label='Renormalized Dispersion (SC)')
plt.axhline(y=delta_sc, color='gray', linestyle=':', label='Superconducting Gap')
plt.xlabel('Momentum (k)')
plt.ylabel('Energy (E)')
plt.title('ARPES Dispersion in Superconductors with Multiple Interactions')
plt.legend()
plt.grid(True)
plt.show()
```

# 4 different models of electronic dispersion
free electrons, kink due to electron-phonon interaction, renormalized bands due to electron-electron interaction, and spin splitting due to spin-orbit coupling
1. **Free Electrons:**
   - Represents the simplest case where electrons behave as free particles with a linear dispersion relation.

2. **Electron-Phonon Interaction (Kink):**
   - The interaction with phonons introduces a kink in the dispersion curve. The self-energy due to electron-phonon coupling is calculated and added to the bare dispersion.

3. **Electron-Electron Interaction (Renormalized Bands):**
   - Electron-electron interactions lead to a renormalization of the band structure, causing a change in the dispersion relation. The self-energy due to electron-electron interactions is included to show this effect.

4. **Spin-Orbit Coupling (Spin Splitting):**
   - Spin-orbit coupling splits the energy bands for spin-up and spin-down electrons. This is modeled by adding a self-energy term that depends on the sign and magnitude of momentum.

```python
import numpy as np
import matplotlib.pyplot as plt

# Define constants
hbar = 1.0545718e-34  # Planck's constant [J*s]
eV = 1.60218e-19      # Electron volt [J]
k_B = 1.380649e-23    # Boltzmann constant [J/K]

# Parameters for the simulation
omega_ph = 0.05       # Phonon energy [eV]
lambda_ph = 1.0       # Electron-phonon coupling constant
lambda_ee = 0.1       # Electron-electron coupling constant (simplified)
lambda_soc = 0.02     # Spin-orbit coupling constant
v_F = 2.0             # Fermi velocity [eV * Å]
k_range = np.linspace(-1.0, 1.0, 500)  # Momentum range [1/Å]

# Bare electron dispersion (linear)
def bare_dispersion(k, v_F):
    return v_F * k

# Self-energy due to electron-phonon interaction (real part)
def self_energy_ph(k, omega_ph, lambda_ph):
    omega = bare_dispersion(k, v_F)
    return -lambda_ph * (omega_ph ** 2) / (omega ** 2 + omega_ph ** 2)

# Self-energy due to electron-electron interaction (simplified)
def self_energy_ee(k, lambda_ee):
    return -lambda_ee * bare_dispersion(k, v_F) ** 2

# Self-energy due to spin-orbit coupling (simplified)
def self_energy_soc(k, lambda_soc):
    return lambda_soc * np.sign(k) * np.sqrt(np.abs(k))

# Renormalized dispersion due to electron-phonon interaction
def renormalized_dispersion_ph(k, v_F, omega_ph, lambda_ph):
    return bare_dispersion(k, v_F) + self_energy_ph(k, omega_ph, lambda_ph)

# Renormalized dispersion due to electron-electron interaction
def renormalized_dispersion_ee(k, v_F, lambda_ee):
    return bare_dispersion(k, v_F) + self_energy_ee(k, lambda_ee)

# Renormalized dispersion due to spin-orbit coupling
def renormalized_dispersion_soc(k, v_F, lambda_soc):
    epsilon_k = bare_dispersion(k, v_F)
    return epsilon_k + self_energy_soc(k, lambda_soc), epsilon_k - self_energy_soc(k, lambda_soc)

# Calculate dispersions
bare_disp = bare_dispersion(k_range, v_F)
renorm_disp_ph = renormalized_dispersion_ph(k_range, v_F, omega_ph, lambda_ph)
renorm_disp_ee = renormalized_dispersion_ee(k_range, v_F, lambda_ee)
renorm_disp_soc_up, renorm_disp_soc_down = renormalized_dispersion_soc(k_range, v_F, lambda_soc)

# Plot the results
fig, axs = plt.subplots(2, 2, figsize=(12, 10))

# Plot 1: Free electrons
axs[0, 0].plot(k_range, bare_disp, label='Free Electrons')
axs[0, 0].set_title('Free Electrons')
axs[0, 0].set_xlabel('Momentum (k)')
axs[0, 0].set_ylabel('Energy (E)')
axs[0, 0].legend()
axs[0, 0].grid(True)

# Plot 2: Electron-Phonon Interaction (Kink)
axs[0, 1].plot(k_range, bare_disp, label='Bare Dispersion', linestyle='--')
axs[0, 1].plot(k_range, renorm_disp_ph, label='With Kink')
axs[0, 1].set_title('Kink due to Electron-Phonon Interaction')
axs[0, 1].set_xlabel('Momentum (k)')
axs[0, 1].set_ylabel('Energy (E)')
axs[0, 1].legend()
axs[0, 1].grid(True)

# Plot 3: Electron-Electron Interaction (Renormalized Bands)
axs[1, 0].plot(k_range, bare_disp, label='Bare Dispersion', linestyle='--')
axs[1, 0].plot(k_range, renorm_disp_ee, label='Renormalized Bands')
axs[1, 0].set_title('Renormalized Bands due to Electron-Electron Interaction')
axs[1, 0].set_xlabel('Momentum (k)')
axs[1, 0].set_ylabel('Energy (E)')
axs[1, 0].legend()
axs[1, 0].grid(True)

# Plot 4: Spin-Orbit Coupling (Spin Splitting)
axs[1, 1].plot(k_range, renorm_disp_soc_up, label='Spin Up')
axs[1, 1].plot(k_range, renorm_disp_soc_down, label='Spin Down')
axs[1, 1].set_title('Spin Splitting due to Spin-Orbit Coupling')
axs[1, 1].set_xlabel('Momentum (k)')
axs[1, 1].set_ylabel('Energy (E)')
axs[1, 1].legend()
axs[1, 1].grid(True)

plt.tight_layout()
plt.show()
```

```python
import numpy as np
import matplotlib.pyplot as plt
from scipy.optimize import curve_fit

# Define constants
hbar = 1.0545718e-34  # Planck's constant [J*s]
eV = 1.60218e-19      # Electron volt [J]
k_B = 1.380649e-23    # Boltzmann constant [J/K]

# Parameters for the simulation
omega_ph = 0.05       # Phonon energy [eV]
lambda_ph = 1.0       # Electron-phonon coupling constant
lambda_ee = 0.1       # Electron-electron coupling constant (simplified)
v_F = 2.0             # Fermi velocity [eV * Å]
k_F = 0.7             # Fermi momentum [Å^-1]
delta_k = 0.05        # Width of MDC peak
delta_E = 0.1         # Width of EDC peak
k_range = np.linspace(0.5, 0.9, 500)  # Momentum range [Å^-1]
E_range = np.linspace(-0.3, 0.1, 500) # Energy range [eV]

# Bare electron dispersion (linear)
def bare_dispersion(k, v_F):
    return v_F * (k - k_F)

# Self-energy due to electron-phonon interaction (real part)
def self_energy_ph(k, omega_ph, lambda_ph):
    omega = bare_dispersion(k, v_F)
    return -lambda_ph * (omega_ph ** 2) / (omega ** 2 + omega_ph ** 2)

# Self-energy due to electron-electron interaction (simplified)
def self_energy_ee(k, lambda_ee):
    return -lambda_ee * bare_dispersion(k, v_F) ** 2

# Total renormalized dispersion
def renormalized_dispersion(k, v_F, omega_ph, lambda_ph, lambda_ee):
    epsilon_k = bare_dispersion(k, v_F)
    self_energy = self_energy_ph(k, omega_ph, lambda_ph) + self_energy_ee(k, lambda_ee)
    return epsilon_k + self_energy

# Generate synthetic ARPES intensity
def arpes_intensity(k, E, v_F, omega_ph, lambda_ph, lambda_ee, delta_k, delta_E):
    epsilon_k = renormalized_dispersion(k, v_F, omega_ph, lambda_ph, lambda_ee)
    intensity = np.exp(-(E - epsilon_k)**2 / (2 * delta_E**2)) * np.exp(-(k - k_F)**2 / (2 * delta_k**2))
    return intensity

# Generate ARPES intensity data
I_k_E = np.array([[arpes_intensity(k, E, v_F, omega_ph, lambda_ph, lambda_ee, delta_k, delta_E) for k in k_range] for E in E_range])

# Extract MDC and EDC
E_cut = 0.0  # Energy cut for MDC
k_cut = 0.7  # Momentum cut for EDC

# MDC at E_cut
MDC = I_k_E[np.argmin(np.abs(E_range - E_cut)), :]
# EDC at k_cut
EDC = I_k_E[:, np.argmin(np.abs(k_range - k_cut))]

# Fit MDC to extract dispersion
def mdc_model(k, k_F, delta_k, A):
    return A * np.exp(-(k - k_F)**2 / (2 * delta_k**2))

popt, _ = curve_fit(mdc_model, k_range, MDC, p0=[k_F, delta_k, max(MDC)])
fitted_dispersion = popt[0]

# Fit EDC to extract self-energy components
def edc_model(E, E_0, delta_E, A):
    return A * np.exp(-(E - E_0)**2 / (2 * delta_E**2))

popt_E, _ = curve_fit(edc_model, E_range, EDC, p0=[bare_dispersion(k_cut, v_F), delta_E, max(EDC)])
fitted_energy = popt_E[0]

# Calculate real and imaginary parts of self-energy
real_self_energy = fitted_energy - bare_dispersion(k_cut, v_F)
imag_self_energy = delta_E

# Plot the results
fig, axs = plt.subplots(1, 2, figsize=(12, 5))

# (a) ARPES intensity with MDC and EDC
axs[0].imshow(I_k_E, extent=[k_range.min(), k_range.max(), E_range.min(), E_range.max()], aspect='auto', origin='lower', cmap='gray')
axs[0].plot([k_F, k_F], [E_range.min(), E_range.max()], 'r--', label='MDC Cut')
axs[0].plot([k_range.min(), k_range.max()], [E_cut, E_cut], 'b--', label='EDC Cut')
axs[0].set_title('ARPES Intensity with MDC and EDC')
axs[0].set_xlabel('Momentum (k)')
axs[0].set_ylabel('Energy (E)')
axs[0].legend()

# (b) Self-energy analysis
E_fit_range = np.linspace(-0.3, 0.0, 300)
real_self_energy_fit = -lambda_ph * (omega_ph ** 2) / (E_fit_range ** 2 + omega_ph ** 2) - lambda_ee * E_fit_range ** 2

axs[1].plot(E_fit_range, real_self_energy_fit, 'r--', label='Model: e-ph + e-e')
axs[1].plot(E_range, I_k_E[:, np.argmin(np.abs(k_range - fitted_dispersion))], 'g', label='ReΣ')
axs[1].axhline(0, color='k', linestyle='--')
axs[1].set_title('Self-Energy Analysis')
axs[1].set_xlabel('Energy (E)')
axs[1].set_ylabel('Σ (eV)')
axs[1].legend()

plt.tight_layout()
plt.show()
```

   - **Real Part (ReΣ):** Extracted from the EDC fitting and compared to the theoretical model for both spin-up and spin-down states.
   - **Imaginary Part (ImΣ):** Estimated from the width of the EDC peak for both spin states.

```python
from matplotlib.animation import FuncAnimation
import numpy as np
import matplotlib.pyplot as plt

# Sample data for self-energy components (e.g., from Fig. 6)
E = np.linspace(-0.3, 0.0, 100)
Re_Sigma = 0.1 * np.sin(np.pi * E)
Im_Sigma = -0.2 * np.exp(-5 * (E + 0.1)**2)  # Simplified example

# Calculate spin lifetime (tau_s)
tau_s = 1.0 / np.abs(Im_Sigma)

# Plot self-energy components and spin lifetime
fig, ax1 = plt.subplots()

ax1.set_xlabel('Energy (E - E_F) (eV)')
ax1.set_ylabel('Self-Energy (eV)', color='tab:blue')
ax1.plot(E, Re_Sigma, 'b--', label='ReΣ')
ax1.plot(E, Im_Sigma, 'b-', label='ImΣ')
ax1.tick_params(axis='y', labelcolor='tab:blue')

ax2 = ax1.twinx()
ax2.set_ylabel('Spin Lifetime (τ_s)', color='tab:red')
ax2.plot(E, tau_s, 'r-', label='τ_s')
ax2.tick_params(axis='y', labelcolor='tab:red')

fig.tight_layout()
plt.title('Self-Energy Components and Spin Lifetime')
plt.show()

# Simulate spin diffusion
grid_size = 100
num_steps = 200
spin_grid = np.ones((grid_size, grid_size))

def update_spin_grid(spin_grid, tau_s, damping_factor=0.01):
    new_spin_grid = spin_grid.copy()
    for i in range(1, grid_size - 1):
        for j in range(1, grid_size - 1):
            neighbors = (spin_grid[i-1, j] + spin_grid[i+1, j] +
                         spin_grid[i, j-1] + spin_grid[i, j+1]) / 4
            new_spin_grid[i, j] = (1 - damping_factor) * spin_grid[i, j] + damping_factor * neighbors
    return new_spin_grid

# Initialize with a spin wave
center = grid_size // 2
for i in range(grid_size):
    for j in range(grid_size):
        distance = np.sqrt((i - center) ** 2 + (j - center) ** 2)
        spin_grid[i, j] = np.cos(2 * np.pi * distance / 10)

# Animate spin diffusion
fig, ax = plt.subplots()

def animate(frame):
    global spin_grid
    spin_grid = update_spin_grid(spin_grid, tau_s, damping_factor=0.01)
    ax.clear()
    ax.imshow(spin_grid, cmap='coolwarm', vmin=-1, vmax=1)
    ax.set_title(f'Step {frame}')

ani = FuncAnimation(fig, animate, frames=num_steps, interval=50)

# To save as GIF
ani.save('spin_diffusion.gif', writer='pillow')

# Alternatively, to save as MP4 video (requires ffmpeg or ImageMagick)
# ani.save('spin_diffusion.mp4', writer='ffmpeg')

plt.show()
```

# Background

*Electron-Electron and Electron-Phonon Interactions, Quasiparticles, and Collective Excitations*

## 1. Basics of Electron-Electron Interactions

### Coulomb Interaction

- **Nature:** Electrons interact via the Coulomb force, which is repulsive due to their negative charge.
- **Hamiltonian:** The Hamiltonian for electron-electron interactions is given by:
  $\
  H_{e-e} = \frac{1}{2} \sum_{i \neq j} \frac{e^2}{|\mathbf{r}_i - \mathbf{r}_j|}$
  where $ \mathbf{r}_i $ and $ \mathbf{r}_j $ are the positions of electrons $ i $ and $ j $.

### Screening

- **Effect:** In a solid, the Coulomb interaction is screened by other electrons, leading to an effective interaction that decays more rapidly with distance.
- **Dielectric Function:** The screening effect can be described using the dielectric function $ \epsilon(\mathbf{q}, \omega) $, which modifies the Coulomb potential.

### Quasiparticles

- **Definition:** Electrons in a solid behave as if they are dressed by interactions with other electrons, forming quasiparticles with modified properties such as effective mass and lifetime.
- **Fermi Liquid Theory:** In metals and normal-state superconductors, the low-energy excitations are quasiparticles that obey Fermi liquid theory. The self-energy $ \Sigma(\mathbf{k}, \omega) $ describes the interaction effects on quasiparticles.

### Many-Body Perturbation Theory

- **GW Approximation:** A common method for treating electron-electron interactions is the GW approximation, where $ G $ is the Green's function and $ W $ is the screened Coulomb interaction.
  $
  \Sigma(\mathbf{k}, \omega) = i \int \frac{d^3q d\omega'}{(2\pi)^4} G(\mathbf{k} - \mathbf{q}, \omega - \omega') W(\mathbf{q}, \omega')
  $

## 2. Basics of Electron-Phonon Interactions

### Phonons

- **Nature:** Phonons are quantized vibrations of the crystal lattice and can be thought of as quasiparticles themselves.
- **Hamiltonian:** The Hamiltonian for phonons in a harmonic approximation is:
  $
  H_{ph} = \sum_{\mathbf{q}} \hbar \omega_{\mathbf{q}} \left( b_{\mathbf{q}}^\dagger b_{\mathbf{q}} + \frac{1}{2} \right)
  $
  where $ b_{\mathbf{q}}^\dagger $ and $ b_{\mathbf{q}} $ are the creation and annihilation operators for a phonon with wave vector $ \mathbf{q} $.

### Electron-Phonon Coupling

- **Interaction:** Electrons interact with phonons, leading to scattering processes that affect their motion.
- **Hamiltonian:** The electron-phonon interaction Hamiltonian is typically written as:
  $
  H_{e-ph} = \sum_{\mathbf{k}, \mathbf{q}, \sigma} g_{\mathbf{q}} \left( b_{\mathbf{q}} + b_{-\mathbf{q}}^\dagger \right) c_{\mathbf{k}+\mathbf{q}, \sigma}^\dagger c_{\mathbf{k}, \sigma}
  $
  where $ g_{\mathbf{q}} $ is the electron-phonon coupling constant, and $ c_{\mathbf{k}, \sigma}^\dagger $ and $ c_{\mathbf{k}, \sigma} $ are the electron creation and annihilation operators.

### Effects on Quasiparticles

- **Mass Renormalization:** Electron-phonon interactions can lead to a renormalization of the electron mass, making quasiparticles heavier than bare electrons.
- **Kinks in Dispersion:** In ARPES spectra, electron-phonon interactions often manifest as kinks in the electronic dispersion curves.

## 3. Quasiparticles and Collective Excitations

### Quasiparticles

- **Concept:** Quasiparticles are emergent phenomena that behave like particles but include the effects of interactions.
- **Effective Mass:** The effective mass of a quasiparticle can be significantly different from the bare electron mass due to interaction effects.

### Collective Excitations

- **Plasmons:** Collective oscillations of the electron density in a material. Plasmons can be excited by electromagnetic fields and are important in optical properties.
- **Magnons:** Collective excitations of the spin system in a magnetically ordered material. They are quantized spin waves.
- **Cooper Pairs:** In superconductors, electrons form Cooper pairs due to attractive interactions mediated by phonons. These pairs condense into a collective ground state that exhibits superconductivity.

## 4. Practical Applications and Implications

### ARPES and Quasiparticles

- **Technique:** ARPES is a powerful technique for studying quasiparticles and their interactions by measuring the energy and momentum of electrons ejected from a material.
- **Dispersion Curves:** ARPES can directly measure the electronic dispersion relation and identify kinks, band renormalization, and spin splitting due to various interactions.

### Spin-Orbit Coupling

- **Effect:** Spin-orbit coupling (SOC) splits the electronic bands based on spin, leading to phenomena such as the Rashba effect in 2D materials and topological surface states.
- **Measurement:** Spin-resolved ARPES (SARPES) can measure the spin polarization of electronic states and provide insights into SOC effects.

# Answer
## 1.1 E-E interaction

#### Self-Energy for Electron-Electron Interactions

The self-energy $\Sigma(\mathbf{k}, \omega)$ is a complex function that describes the many-body effects on an electron's propagation in a material. It consists of a real part $\Sigma'(\mathbf{k}, \omega)$ and an imaginary part $\Sigma''(\mathbf{k}, \omega)$.

1. **Real Part ($\Sigma'(\mathbf{k}, \omega)$):** 
   - This term accounts for the energy shift of the quasiparticle due to interactions.
   - It modifies the effective mass and dispersion relation of the electrons.

2. **Imaginary Part ($\Sigma''(\mathbf{k}, \omega)$):**
   - This term describes the scattering rate and lifetime of the quasiparticles.
   - It is related to the broadening of spectral features in ARPES.

The self-energy for electron-electron interactions can be expressed using the GW approximation or other many-body perturbation theories. In the GW approximation, the self-energy is given by:

$
\Sigma(\mathbf{k}, \omega) = i \int \frac{d\omega'}{2\pi} \sum_{\mathbf{q}} G(\mathbf{k} + \mathbf{q}, \omega + \omega') W(\mathbf{q}, \omega')
$

Where:
- $G(\mathbf{k}, \omega)$ is the Green's function.
- $W(\mathbf{q}, \omega)$ is the dynamically screened Coulomb interaction.

#### Effects on the Dispersion Curve and Observed Kinks

Electron-electron interactions significantly affect the electronic dispersion observed in ARPES experiments.

1. **Renormalization of the Dispersion:**
   - The real part of the self-energy $\Sigma'(\mathbf{k}, \omega)$ modifies the bare dispersion relation $\epsilon_{\mathbf{k}}$. The observed dispersion relation becomes $E(\mathbf{k}) = \epsilon_{\mathbf{k}} + \Sigma'(\mathbf{k}, \omega)$.
   - This renormalization can lead to changes in the effective mass of the electrons, observed as a modification of the slope of the dispersion curve near the Fermi level.

2. **Kinks in the Dispersion:**
   - Kinks in the dispersion curve are often observed in ARPES data. These are sharp changes in the slope of the dispersion relation.
   - They can arise from strong electron-electron interactions, particularly when coupled with collective excitations such as plasmons or phonons.
   - The energy at which a kink appears corresponds to the energy of these collective modes.

3. **Lifetime and Broadening:**
   - The imaginary part $\Sigma''(\mathbf{k}, \omega)$ affects the linewidth and lifetime of the quasiparticle states.
   - Higher values of $\Sigma''(\mathbf{k}, \omega)$ lead to broader peaks in ARPES spectra, indicating shorter lifetimes and higher scattering rates.

#### Observing and Interpreting ARPES Data

- **Quasiparticle Peaks:**
  - Sharp peaks in ARPES spectra correspond to well-defined quasiparticle states with longer lifetimes.
  - Broadened peaks indicate significant electron-electron scattering.

- **Energy and Momentum Dependence:**
  - By analyzing the energy and momentum dependence of the spectral function $\A(\mathbf{k}, \omega)$, one can extract the self-energy components.
  - The spectral function is related to the Green's function as $\A(\mathbf{k}, \omega) = -\frac{1}{\pi} \text{Im} G(\mathbf{k}, \omega)$.

## 1.2 SOC

### Appearance of Kinks

1. **Renormalization Effects:**
   - Similar to electron-phonon interactions, SOC can lead to renormalization effects in the band structure, potentially contributing to the appearance of kinks in the dispersion curves.
   - These kinks can arise from the interaction of the electron's spin with the lattice or other spins, leading to a sudden change in the slope of the dispersion relation.

2. **Spin-Dependent Kinks:**
   - Kinks influenced by SOC might be spin-dependent, meaning they could manifest differently for spin-up and spin-down electrons, leading to observable spin-polarized features in ARPES measurements.

### Modeling Spin-Orbit Coupling

SOC can be modeled similarly to electron-phonon interactions, but with distinct differences due to its intrinsic nature. Here's how SOC can be incorporated into theoretical models:

1. **Hamiltonian Approach:**
   - The SOC is often included in the Hamiltonian of the system. A common form of the SOC term in the Hamiltonian is:
     $ \H_{\text{SOC}} = \lambda \mathbf{L} \cdot \mathbf{S}$
     where $\lambda$ is the SOC strength, $\mathbf{L}$ is the orbital angular momentum, and $\mathbf{S}$ is the spin angular momentum.

2. **Spin-Dependent Self-Energy:**
   - The self-energy can be modified to include spin-dependent terms. The total self-energy $\Sigma(\mathbf{k}, \omega)$ can be decomposed into spin-up and spin-down components:
     $ \Sigma(\mathbf{k}, \omega) = \Sigma_{\uparrow}(\mathbf{k}, \omega) \sigma_{\uparrow} + \Sigma_{\downarrow}(\mathbf{k}, \omega) \sigma_{\downarrow}$
     where $\sigma_{\uparrow}$ and $\sigma_{\downarrow}$ are spin projection operators.

3. **Dynamical Mean-Field Theory (DMFT):**
   - Advanced many-body techniques such as DMFT can incorporate SOC effects by solving the self-consistent equations that include SOC terms. This approach provides a more accurate description of the electronic structure in strongly correlated systems.

4. **First-Principles Calculations:**
   - Density functional theory (DFT) and related first-principles methods can include SOC by explicitly adding the SOC term to the Kohn-Sham equations. This approach can yield detailed information about the band structure and SOC-induced splitting.

# 1.3 Multiple interaction

### Combining Self-Energies

The total self-energy $\Sigma(\mathbf{k}, \omega)$ is a combination of contributions from different interactions. Each type of interaction has its characteristic self-energy, and these can be added to form the total self-energy:

$
\Sigma(\mathbf{k}, \omega) = \Sigma_{\text{el-el}}(\mathbf{k}, \omega) + \Sigma_{\text{el-ph}}(\mathbf{k}, \omega) + \Sigma_{\text{SOC}}(\mathbf{k}, \omega)
$

1. **Electron-Electron Interaction ($\Sigma_{\text{el-el}}(\mathbf{k}, \omega)$):**
   - Describes the many-body effects due to electron-electron Coulomb interactions.
   - Typically calculated using many-body perturbation theory (e.g., GW approximation).

2. **Electron-Phonon Interaction ($\Sigma_{\text{el-ph}}(\mathbf{k}, \omega)$):**
   - Accounts for the interaction of electrons with lattice vibrations (phonons).
   - Often modeled using the Migdal-Eliashberg theory, which describes how phonons mediate electron pairing in superconductors.

3. **Spin-Orbit Coupling ($\Sigma_{\text{SOC}}(\mathbf{k}, \omega)$):**
   - Describes the interaction between an electron’s spin and its motion around the nucleus.
   - Incorporated through relativistic extensions of the Hamiltonian, affecting both the band structure and scattering rates.

#### Methodologies for Integration

1. **Perturbation Theory:**
   - Use perturbation theory to combine different interaction terms. The total self-energy is obtained by summing the contributions from each interaction within the perturbative framework.

2. **Dynamical Mean-Field Theory (DMFT):**
   - DMFT can handle strong correlations and can be extended to include electron-phonon and SOC effects. The DMFT self-consistent loop is used to compute the total self-energy.

3. **First-Principles Methods:**
   - Density Functional Theory (DFT) and GW calculations can be extended to include SOC and electron-phonon interactions. These methods provide a quantitative description of the electronic structure.

4. **Effective Hamiltonians:**
   - Construct effective Hamiltonians that include terms representing each interaction. Diagonalize the Hamiltonian to obtain the electronic states and their energies.

#### Dominant Interactions and Their Interplay

The dominant interactions in a material depend on its specific properties, such as atomic composition, crystal structure, and temperature. Here's how these interactions interplay:

1. **Electron-Electron vs. Electron-Phonon Interactions:**
   - In metals, electron-phonon interactions often dominate at low energies, leading to superconductivity.
   - In strongly correlated materials (e.g., Mott insulators), electron-electron interactions are dominant and can lead to phenomena like magnetism and metal-insulator transitions.

2. **Electron-Electron Interactions:**
   - Lead to correlation effects such as band renormalization and quasiparticle lifetime reduction.
   - Cause phenomena like the formation of Hubbard bands and spin fluctuations.

3. **Electron-Phonon Interactions:**
   - Cause renormalization of the electronic band structure and can introduce kinks in the dispersion.
   - Responsible for the formation of Cooper pairs in conventional superconductors.

4. **Spin-Orbit Coupling:**
   - Induces band splitting and modifies the dispersion, particularly in materials with heavy elements.
   - Can lead to topologically protected states, such as those in topological insulators.

5. **Interplay of Interactions:**
   - SOC can enhance or compete with electron-electron interactions, affecting the magnetic and electronic properties.
   - Electron-phonon interactions can be modified by the presence of strong electron-electron interactions, as in high-temperature superconductors.
   - The combined effect of these interactions can lead to complex phases, such as spin-density waves, charge-density waves, and unconventional superconductivity.

## 1.3 Temperature effects
#### Temperature Dependence of Interactions

1. **Electron-Phonon Interaction:**
   - **Phonon Population:**
     - As temperature increases, the population of phonons (lattice vibrations) increases, leading to enhanced electron-phonon scattering.
     - The self-energy term $\Sigma_{\text{el-ph}}(\mathbf{k}, \omega)$ becomes more significant at higher temperatures, leading to greater renormalization and scattering rates.
   - **Debye-Waller Factor:**
     - This factor, which accounts for the thermal motion of atoms, affects the electron-phonon coupling strength.
     - Higher temperatures increase the Debye-Waller factor, which can reduce the intensity of photoemission signals and broaden the spectral features.

2. **Electron-Electron Interaction:**
   - **Screening Effects:**
     - Temperature affects the dielectric function, which in turn modifies the screening of electron-electron interactions.
     - At higher temperatures, increased thermal excitations can enhance screening, reducing the effective electron-electron interaction strength.
   - **Thermal Broadening:**
     - Thermal excitations cause broadening of the Fermi-Dirac distribution, impacting the electron-electron scattering rates.
     - This broadening can lead to an increase in the imaginary part of the self-energy $\Sigma''_{\text{el-el}}(\mathbf{k}, \omega)$, resulting in shorter quasiparticle lifetimes and broader spectral features.

3. **Spin-Orbit Coupling:**
   - **Indirect Effects:**
     - SOC itself is relatively temperature-independent, but the overall effect of SOC on the electronic structure can be influenced by temperature-dependent changes in electron-electron and electron-phonon interactions.
   - **Thermal Fluctuations:**
     - Thermal fluctuations can indirectly affect SOC-induced features by modifying the overall band structure and scattering processes.

#### Including Temperature Effects in the Model

1. **Self-Energy Modifications:**
   - Incorporate temperature-dependent self-energy terms for electron-phonon ($\Sigma_{\text{el-ph}}(T, \mathbf{k}, \omega)$) and electron-electron ($\Sigma_{\text{el-el}}(T, \mathbf{k}, \omega)$) interactions.
   - These can be computed using finite-temperature many-body perturbation theories or first-principles methods that include temperature effects.

2. **Phonon Spectral Function:**
   - The phonon spectral function $B(\mathbf{q}, \omega, T)$ describes the phonon density of states and depends on temperature.
   - The electron-phonon self-energy at finite temperature can be written as:
     $     \Sigma_{\text{el-ph}}(T, \mathbf{k}, \omega) = \sum_{\mathbf{q}, \nu} |g_{\mathbf{k}, \mathbf{q}, \nu}|^2 \left[ \frac{n_B(\omega_{\mathbf{q}, \nu}, T) + 1 - f(\epsilon_{\mathbf{k}+\mathbf{q}}, T)}{\omega - \epsilon_{\mathbf{k}+\mathbf{q}} + \omega_{\mathbf{q}, \nu} + i\eta} + \frac{n_B(\omega_{\mathbf{q}, \nu}, T) + f(\epsilon_{\mathbf{k}+\mathbf{q}}, T)}{\omega - \epsilon_{\mathbf{k}+\mathbf{q}} - \omega_{\mathbf{q}, \nu} + i\eta} \right]$
     where $\n_B(\omega, T)$ is the Bose-Einstein distribution function for phonons and $\f(\epsilon, T)$ is the Fermi-Dirac distribution function for electrons.

3. **Quasiparticle Lifetime:**
   - Calculate the temperature-dependent imaginary part of the self-energy ($\Sigma''(T, \mathbf{k}, \omega)$) to account for the broadening of spectral features.
   - The broadening at finite temperatures can be included as:
     $
     \Gamma(T, \mathbf{k}) = -2 \Sigma''(T, \mathbf{k}, \omega)
     $
     where $\Gamma$ represents the quasiparticle scattering rate.

4. **Spectral Function:**
   - The spectral function $\A(T, \mathbf{k}, \omega)$ at finite temperatures can be obtained using the temperature-dependent self-energy:
     $ \A(T, \mathbf{k}, \omega) = -\frac{1}{\pi} \frac{\Sigma''(T, \mathbf{k}, \omega)}{\left[\omega - \epsilon_{\mathbf{k}} - \Sigma'(T, \mathbf{k}, \omega)\right]^2 + \left[\Sigma''(T, \mathbf{k}, \omega)\right]^2}$

#### Role of Temperature in Broadening and Kinks

1. **Broadening of Spectral Features:**
   - As temperature increases, the imaginary part of the self-energy ($\Sigma''(T, \mathbf{k}, \omega)$) increases, leading to broader spectral features in ARPES.
   - This broadening reflects shorter quasiparticle lifetimes and increased scattering rates due to thermal excitations.

2. **Modification of Kinks:**
   - Kinks in the dispersion curves, often associated with strong electron-phonon coupling, can be smeared out at higher temperatures due to increased phonon population and enhanced electron-phonon scattering.
   - The temperature dependence of kinks can provide insights into the coupling strength and the nature of the interactions involved.
